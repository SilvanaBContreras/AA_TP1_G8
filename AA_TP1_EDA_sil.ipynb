## Feat engineering

Crearía:

-ratio Unidades_curriculares_1er_semestre_(aprobadas)
sobre Unidades_curriculares_1er_semestre_(inscrito).

-ratio Unidades_curriculares_2do_semestre_(aprobadas) Unidades_curriculares_2do_semestre_(inscrito)

-producto Nota_promedio_en_el_2do_semestre * Unidades_curriculares_2do_semestre_(inscrito)

-producto Nota_promedio_en_el_1er_semestre * Unidades_curriculares_1er_semestre_(inscrito)

-lag Nota_promedio_en_el_2do_semestre vs Nota_promedio_en_el_1er_semestre

```
# Ratio 1er semestre
dataset['ratio_aprobadas_inscrito_1sem_fe'] = (
    dataset['Unidades_curriculares_1er_semestre_(aprobadas)'] /
    (dataset['Unidades_curriculares_1er_semestre_(inscrito)'] + 0.001)
)

# Ratio 2do semestre
dataset['ratio_aprobadas_inscrito_2sem_fe'] = (
    dataset['Unidades_curriculares_2do_semestre_(aprobadas)'] /
    (dataset['Unidades_curriculares_2do_semestre_(inscrito)'] + 0.001)
)

# Apago porque da muy correlacionada
# # Producto 1er semestre
# dataset['producto_nota_inscritos_1sem_fe'] = dataset['Nota_promedio_en_el_1er_semestre'] * dataset['Unidades_curriculares_1er_semestre_(inscrito)']

# # Producto 2do semestre
# dataset['producto_nota_inscritos_2sem_fe'] = dataset['Nota_promedio_en_el_2do_semestre'] * dataset['Unidades_curriculares_2do_semestre_(inscrito)']

#Lag notas promedio
dataset['lag_nota_promedio_fe'] = dataset['Nota_promedio_en_el_2do_semestre'] - dataset['Nota_promedio_en_el_1er_semestre']

```

## X e y





```
X = dataset.drop(['target_label', 'target_binaria'], axis=1)
y = dataset['target_binaria']
```

## tipos transformación

```
X_transf=X.copy()
```

```
# Estado Civil achicamos niveles

mapeo_agrupaciones = {
    '5 - Unión de hecho': '2 - Casado/Unión de Hecho',
    '2 - Casado': '2 - Casado/Unión de Hecho',
    '6 - Separado legalmente': '4 - Divorciado/Separado',
    '4 - Divorciado': '4 - Divorciado/Separado',
}

X_transf['estado_civil_label'] = X_transf['estado_civil_label'].replace(mapeo_agrupaciones)


# 3. Verificar los nuevos valores (opcional)
print("Nuevos valores únicos en la columna 'estado_civil_label':")
print(X_transf['estado_civil_label'].unique())
```

```
cat_cols =[]
num_cols = []
bin_cols=[]
for c in X.columns:
  if X[c].nunique()<= 2:
    bin_cols.append(c)
  elif X[c].nunique() <=20:
    cat_cols.append(c)
  else:
    num_cols.append(c)
ord_cols=['Estudios_máximos_antes_de_la_inscripción_label','Demanda_Cog_Padre_label','Demanda_Cog_Madre_label','Ingresos_Padre_label','Ingresos_Madre_label']
nom_cols=[c for c in cat_cols if c not in ord_cols]
print(f'Total de variables : {X.shape[1]}')
print(f'   - variables binarias : {len(bin_cols)}')
print(f'   - variables numericas : {len(num_cols)}')
print(f'   - variables categoricas : {len(cat_cols)}')
print(f'        *variables Nominales : {len(nom_cols)}')
print(f'        *variables Ordinales : {len(ord_cols)}')
```

```


#transformación binarias a numéricas
for c in bin_cols:
  X_transf[f'{c}_transformed'] = X_transf[c].apply(lambda x : 0 if x[0]=='0' else 1 )
```

```
estudios_niveles = {'Otros / No completado':0,
           'Educación Básica':1,
    'Educación Secundaria':2,
    'Superior - Pregrado':3,
    'Superior - Posgrado':4
}
cognitivos_niveles={'Muy alta demanda cognitiva':5,'Alta demanda cognitiva':4,
 'Muy baja demanda cognitiva':1 ,'Demanda cognitiva baja':2,'Demanda cognitiva media':3 }

ingresos_niveles={'Ingresos medios-bajos':2, 'Altos ingresos':5, 'Bajos ingresos':1,
 'Ingresos medios-altos':4, 'Ingresos medios':3}

#transformación ordinales a numéricas
X_transf["Estudios_máximos_antes_de_la_inscripción_label_transformed"] = X_transf['Estudios_máximos_antes_de_la_inscripción_label'].apply(lambda x : estudios_niveles[x])
X_transf["Demanda_Cog_Padre_label_transformed"]=X["Demanda_Cog_Padre_label"].apply(lambda x: cognitivos_niveles[x])
X_transf["Demanda_Cog_Madre_label_transformed"]=X["Demanda_Cog_Madre_label"].apply(lambda x: cognitivos_niveles[x])
X_transf["Ingresos_Padre_label_transformed"]=X["Ingresos_Padre_label"].apply(lambda x: ingresos_niveles[x])
X_transf["Ingresos_Madre_label_transformed"]=X["Ingresos_Madre_label"].apply(lambda x: ingresos_niveles[x])
```

```
#transformación nominales a numéricas (dummies)

X_transf=pd.get_dummies(X_transf,columns=nom_cols,dtype=int)
```

```
#Dropeo columnas strings

X_transf = X_transf.drop(columns=ord_cols+bin_cols)
```

## correlaciones entre numéricas

```
correlaciones= X_transf[num_cols].corr()
correlaciones[np.abs(correlaciones) <0.8] =0
correlaciones
```

```
sns.heatmap(correlaciones)
```

```
np.where(correlaciones>0)
```

```
# Convertir la matriz a una Series y luego resetear el índice para crear un DataFrame
correlaciones_pares = correlaciones.unstack().reset_index()

# Renombrar las columnas
correlaciones_pares.columns = ['Variable 1', 'Variable 2', 'Correlacion']
```

```
# Aplicar los filtros
pares_fuertes = correlaciones_pares[
    (correlaciones_pares['Correlacion'] != 0.0) &
    (correlaciones_pares['Correlacion'] != 1.0) &
    (correlaciones_pares['Variable 1'] < correlaciones_pares['Variable 2'])
]

pares_fuertes = pares_fuertes.sort_values(by='Correlacion', ascending=False)

print("\nPares de columnas con correlación |r| >= 0.7:")
print(pares_fuertes)
```

```
columnas_a_eliminar = []

columnas_a_eliminar.append('Unidades_curriculares_2do_semestre_(inscrito)')
columnas_a_eliminar.append('Unidades_curriculares_2do_semestre_(aprobadas)')

# Mostrar el vector final
print("\nVector de columnas para eliminar:")
print(columnas_a_eliminar)
```

## X_transf_short creación

```
X_transf_short = X_transf.drop(
    columns=columnas_a_eliminar,
    axis=1,
    inplace=False
)

print(f"Columnas originales: {X_transf.shape[1]}")
print(f"Columnas eliminadas: {len(columnas_a_eliminar)}")
print(f"Columnas finales en X_transf_final: {X_transf_short.shape[1]}")
```

# train test estratificado


Armo dos opciones de entrenamiento con correlacionadas y sin

```
from sklearn.model_selection import StratifiedKFold , StratifiedShuffleSplit

sss = StratifiedShuffleSplit(n_splits =1 ,test_size=0.3 , random_state=14)
for train_index, test_index in sss.split(X_transf,y):
  X_train=X_transf.loc[train_index]
  y_train = y.loc[train_index]
  X_test = X_transf.loc[test_index]

  X_train_short = X_transf_short.loc[train_index] #Alternativa sin correlacionadas
  X_test_short = X_transf_short.loc[test_index] #Alternativa sin correlaciones

  y_test = y.loc[test_index]
  print(train_index)
```

```
print("Dimensiones del conjunto de características (X):")
print(f"X_train: {X_train.shape}")
print("\nDimensiones del conjunto de características (X) sin correlacionadas:")
print(f"X_train_short: {X_train_short.shape}")
print("\nDimensiones de la variable objetivo (y):")
print(f"X_test: {X_test.shape}")
print("\nDimensiones de la variable objetivo (y) sin correlaciones:")
print(f"X_test_short: {X_test_short.shape}")
print("\nDimensiones de la variable objetivo (y):")
print(f"y_train: {y_train.shape}")
print(f"y_test: {y_test.shape}")
```

## OUTPUT

```
# Lista de conjuntos y sus nombres de archivo
conjuntos_a_guardar = [
    (X_train, 'X_train_completo'),
    (y_train, 'y_train'),  # Estas son Series y causan el error
    (X_test, 'X_test_completo'),
    (y_test, 'y_test'),    # Estas son Series y causan el error
    (X_train_short, 'X_train_reducido'),
    (X_test_short, 'X_test_reducido')
]

for df, nombre in conjuntos_a_guardar:
    # 1. Comprobamos si el objeto es una Series
    if isinstance(df, pd.Series):
        # 2. Si es Series, lo convertimos a un DataFrame de una columna para poder usar .to_parquet()
        df_a_guardar = df.to_frame()
    else:
        # 3. Si ya es un DataFrame, lo usamos directamente
        df_a_guardar = df

    # index=True es importante para conservar los índices de la división de sklearn
    df_a_guardar.to_parquet(f'{nombre}.parquet', index=True)
    print(f"✅ Guardado: {nombre}.parquet")
```
