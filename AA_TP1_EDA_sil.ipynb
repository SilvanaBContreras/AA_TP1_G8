## creo columna target_binaria

```
#target_label ['Desertor' 'En Curso' 'Graduado']

dataset['target_binaria'] = dataset['target_label'].apply(lambda x: 1 if x == 'Desertor' else 0)
```

```
dataset['target_binaria'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribución de target_binaria')
plt.show()
```

```
sns.catplot(x="sexo_label", hue="target_binaria", kind="count", data=dataset, palette="Set1")
plt.xlabel('Sexo')
plt.ylabel('Número de estudiantes')
plt.title('Proporción de deserción por sexo')
plt.show()

sns.catplot(x="Carrera", hue="target_binaria", kind="count", data=dataset, palette="Set1")
plt.xlabel('Carrera')
plt.ylabel('Número de estudiantes')
plt.title('Proporción de deserción por carrera')
plt.xticks(rotation=45, ha='right')  # Opcional: para rotar las etiquetas si son muy largas
plt.show()

sns.catplot(x="estado_civil_label", hue="target_binaria", kind="count", data=dataset, palette="Set1")
plt.xlabel('Estado Civil')
plt.ylabel('Número de estudiantes')
plt.title('Proporción de deserción por estado civil')
plt.show()
```

```
dataset.plot(kind='hist', column='Edad_al_momento_de_la_inscripción', by='target_binaria', figsize=(10, 7))
plt.show()
```

```
dataset.groupby('target_binaria')['Edad_al_momento_de_la_inscripción'].plot(kind='hist', bins=20, alpha=0.7, legend=True)

plt.xlabel('Edad_al_momento_de_la_inscripción')
plt.ylabel('Densidad')
plt.title('Distribución de Edad_al_momento_de_la_inscripción por clase')
```

```
dataset[['target_binaria', 'Edad_al_momento_de_la_inscripción']].plot(kind='box', by='target_binaria')

plt.xlabel('Desertor')
plt.ylabel('Edad_al_momento_de_la_inscripción')
plt.title('Distribución de edades inscripción por target')
```

```
plt.figure(figsize=(8, 6))
sns.boxplot(x='target_binaria', y='Edad_al_momento_de_la_inscripción', data=dataset)

# Establecer las etiquetas de los ejes y el título del gráfico
plt.xlabel('Desertor')
plt.ylabel('Edad_al_momento_de_la_inscripción')
plt.title('Distribución de edades inscripción por target')
```

```
sns.violinplot(x='target_binaria', y='Edad_al_momento_de_la_inscripción', data=dataset)

# establecer las etiquetas de los ejes y el título
plt.xlabel('target_binaria')
plt.ylabel('Edad_al_momento_de_la_inscripción')
plt.title('Violin plot de Edad_al_momento_de_la_inscripción por target')
plt.show()
```

```
dataset.plot(x='Cualificación_promedio_de_estudios_previos', y='Puntaje_en_examen_de_ingreso', kind='scatter')
plt.xlabel('Cualificación_promedio_de_estudios_previos')
plt.ylabel('Puntaje_en_examen_de_ingreso')
plt.title('Gráfico de dispersión de calificación estudios previos vs calificación ingreso')
plt.show()


```

```
colors = {1:'blue', 2:'red'}

plt.subplots(figsize=(8, 6))
sns.scatterplot(data=dataset, x='Cualificación_promedio_de_estudios_previos', y='Puntaje_en_examen_de_ingreso', hue='target_binaria', palette=colors, alpha=0.6)

# definimos las etiquetas de los ejes y el título del gráfico
plt.xlabel('Cualificación_promedio_de_estudios_previos')
plt.ylabel('Puntaje_en_examen_de_ingreso')
plt.title('Gráfico de dispersión ')
plt.show()
```

```
pd.crosstab(dataset['target_binaria'], dataset['sexo_label']).plot(kind='bar', stacked=True)
plt.xlabel('Desertor')
plt.ylabel('Frecuencia')
plt.title('Distribución de desertores por Sexo')
plt.legend(title='Sexo', loc='upper left')
plt.show()
```
